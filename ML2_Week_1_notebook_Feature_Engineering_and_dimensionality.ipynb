## Handling Missing Data

```python
import pandas as pd
data = {
    'Size (sq ft)': [1500, 2500, 1800, 2200, None],
    'Number of Bedrooms': [3, 4, 3, 5, 4],
    'Number of Bathrooms': [2, 3, None, 3, 2],
    'Year Built': [2000, 2010, 2005, None, 2018],
    'Location': ['Suburb', 'City', 'City', 'Suburb', None],
    'Price': [300000, 450000, 350000, 500000, 420000]
}

df = pd.DataFrame(data)
df
```

- Strategies used to handle missing data
    - Removing rows with missing data
    - Impute missing values (Fill with the mean, mode, median, ), Use advamced KNN Imputations

```python
df
```

```python
df.info()
```

### Handling missing values using the imputer from sklearn

```python
# Imputing missing values with mean
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
df['Size (sq ft)'] = imputer.fit_transform(df[['Size (sq ft)']])
df
```

```python
# Imputing missing values with the mean
df['Size (sq ft)'].fillna(df['Size (sq ft)'].mean(), inplace=True)
df
```

```python
# Impute with median
df['Number of Bathrooms'].fillna(df['Number of Bathrooms'].median(), inplace = True)
df
```

```python
# Filling with the mode
df['Location'].fillna(df['Location'].mode()[0], inplace=True)
df
```

```python
df['Year Built'].fillna(df['Year Built'].mode()[0], inplace=True)
df
```

## Feature Engineering
### Creating new Features

```python
df['Size Per Bedroom'] = df['Size (sq ft)'] / df['Number of Bedrooms']
df
```

```python
# Creating a feature age of the house
df['Age of House'] = 2024 - df['Year Built']
df
```

## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

# Label Encoding
label_encoder = LabelEncoder()
df['Location_Label'] = label_encoder.fit_transform(df['Location'])
df
```

### One-Hot Encoding

```python
# One hot encoder
df_one_hot = pd.get_dummies(df, columns=['Location'])

df_one_hot
```
