## Gausian Mixed Models (GMM)
- Type of Probabilistic model used to represent normally distributed subpopulations within an overal population.

### Key Concepts:
1. Mixture Models:
    - It assumes the data comes from a mixture of several probability distributions

2. Gausian Distribution:
    - Each subplot is assumed to be normally distributed

3. Soft Clustering
    - GMM assigns a probability to each of the data points belonging to each cluster

4. Expectation-Maximization (EM) Algorithm
    - GMM parameters are estimated using the EM algorithm. It iteratively improves the likelihood of the data given the model.

### Why Use GMM for Clustering:
- Flexibility
- Soft Clustering
- Multidimensional data

### Steps to Use GMM for clustering
1. Innitialization - Parameters (Means, covariances, and mixture weights)
2. Expectation Step (E-Step) - Calculate the probability (Responsibility) that each data point belongs to each gausian component
3. Maximization Step (M-Step) - Update the parameters of the Gausina compoents to maximize the likelihood of the data Given the responsibilities
4. Convergence - Repeat E-step and M-step till convergence, typically when the change in the log-likelihood between iterations is below a threshold.

```python
# Import the libraries needed
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.mixture import GaussianMixture
```

```python
# Generate a dataset
X, y_true = make_blobs(n_samples=400, centers=4, cluster_std=0.60, random_state=0)
```

```python
# Fit the model to the data
gmm = GaussianMixture(n_components=4)
gmm.fit(X)
labels = gmm.predict(X)
```

```python
# Plot the data with the cluster assignments
plt.scatter(X[:, 0], X[:, 1], c=labels, s=40, cmap='viridis')
plt.title("Gausian Mixture Model Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

## Plot the probability of each point belonging to each cluster

```python
probs = gmm.predict_proba(X)
plt.scatter(X[:, 0], X[:, 1], c=probs.max(axis=1), s = 40, cmap="viridis")
plt.title("Probability of Cluster membership")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.colorbar(label="Probability")
plt.show()
```
